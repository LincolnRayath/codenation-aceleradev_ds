```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train_corr = train.corr()
train_corr.head()
```

```python
treino_corr = train_corr.loc['IN_TREINEIRO']
treino_corr
```

```python
aux = treino_corr[(treino_corr >= 0.25) | (treino_corr <= -0.25)]
aux
```

```python
variaveis_sel = aux.index
variaveis_sel
```

```python
df_train = train[variaveis_sel]
```

```python
y_train = df_train['IN_TREINEIRO']
X_train = df_train.drop(columns = 'IN_TREINEIRO')
```

```python
X_test = test[X_train.columns]
X_test
```

```python
X_train = ss.fit_transform(X_train) 
X_test = ss.transform(X_test)
```

```python
#Instanciando o modelo Random Forest Classifier
classifier = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
#Treinando o modelo
classifier.fit(X_train, y_train)
```

```python
y_pred_test = classifier.predict(X_test)
```

```python
df_answer = pd.DataFrame()
df_answer['NU_INSCRICAO'] = test['NU_INSCRICAO']
df_answer['IN_TREINEIRO'] = y_pred_test
df_answer
```

```python
df_answer.to_csv('answer.csv', index=False, header=True)
```
